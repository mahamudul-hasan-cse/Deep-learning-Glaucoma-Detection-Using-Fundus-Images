```python
!pip install -q segmentation-models-pytorch albumentations timm
!pip install -q numpy==1.26.4 --force-reinstall

# ============================================================
# IMPORTS & SETUP
# ============================================================
import os, glob, random
import numpy as np
import cv2
import albumentations as A
import segmentation_models_pytorch as smp
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
import pandas as pd
from torch.amp import autocast, GradScaler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

def seed_all(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_all(42)

# ============================================================
# PATHS
# ============================================================
REFUGE_ROOT = "/kaggle/input/refuge2/REFUGE2"

def get_refuge_pairs(split: str):
    img_dir  = os.path.join(REFUGE_ROOT, split, "images")
    mask_dir = os.path.join(REFUGE_ROOT, split, "mask")
    imgs  = sorted(glob.glob(os.path.join(img_dir, "*")))
    masks = sorted(glob.glob(os.path.join(mask_dir, "*")))
    print(f"{split}: {len(imgs)} images, {len(masks)} masks")
    return imgs, masks

# ============================================================
# FUNDUS CLEANING
# ============================================================
def clean_fundus(img_rgb):
    try:
        k1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
        k2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
        closed = cv2.morphologyEx(img_rgb, cv2.MORPH_CLOSE, k1)
        closed = cv2.morphologyEx(closed, cv2.MORPH_CLOSE, k2)
        blurred = cv2.GaussianBlur(closed, (5,5), 0)
        bg = cv2.GaussianBlur(blurred, (41,41), 0)
        norm = cv2.add(cv2.subtract(blurred, bg), 128)
        norm = np.clip(norm, 0, 255).astype(np.uint8)
        lab  = cv2.cvtColor(norm, cv2.COLOR_RGB2LAB)
        L, A_, B_ = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.5, tileGridSize=(8,8))
        L2 = clahe.apply(L)
        lab2 = cv2.merge((L2, A_, B_))
        out = cv2.cvtColor(lab2, cv2.COLOR_LAB2RGB)
        return out
    except:
        return img_rgb

# ============================================================
# DATASET
# ============================================================
class RefugeSegDataset(Dataset):
    def __init__(self, img_paths, mask_paths, transforms=None, img_size=512):
        self.img_paths  = img_paths
        self.mask_paths = mask_paths
        self.transforms = transforms
        self.img_size   = img_size

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.img_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_UNCHANGED)
        if mask.ndim == 3:
            mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)

        img  = cv2.resize(img,  (self.img_size, self.img_size))
        mask = cv2.resize(mask, (self.img_size, self.img_size),
                          interpolation=cv2.INTER_NEAREST)

        green   = img[:, :, 1]
        cleaned = clean_fundus(img)
        four_ch = np.concatenate([cleaned, green[..., None]], axis=-1)  # (H,W,4)

        mask_class = np.zeros_like(mask, dtype=np.uint8)
        mask_class[mask == 128] = 1   # disc
        mask_class[mask == 255] = 2   # cup

        if self.transforms:
            aug = self.transforms(image=four_ch, mask=mask_class)
            four_ch   = aug["image"]
            mask_class = aug["mask"]

        four_ch = (four_ch / 255.0).astype(np.float32)
        four_ch = four_ch.transpose(2, 0, 1)  # (4,H,W)

        one_hot = np.zeros((3, self.img_size, self.img_size), dtype=np.float32)
        one_hot[0] = (mask_class == 0)
        one_hot[1] = (mask_class == 1)
        one_hot[2] = (mask_class == 2)

        return torch.tensor(four_ch), torch.tensor(one_hot)

# ============================================================
# AUGMENTATIONS
# ============================================================
train_tf = A.Compose([
    A.Resize(512, 512),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.10,
                       rotate_limit=120, border_mode=cv2.BORDER_CONSTANT, p=0.5),
    A.RandomBrightnessContrast(0.1, 0.1, p=0.5),
])
val_tf = A.Compose([A.Resize(512, 512)])

# ============================================================
# LOAD DATA
# ============================================================
train_imgs, train_masks = get_refuge_pairs("train")
val_imgs,   val_masks   = get_refuge_pairs("val")
test_imgs,  test_masks  = get_refuge_pairs("test")

train_ds = RefugeSegDataset(train_imgs, train_masks, train_tf)
val_ds   = RefugeSegDataset(val_imgs,   val_masks,   val_tf)
test_ds  = RefugeSegDataset(test_imgs,  test_masks,  val_tf)

train_ld = DataLoader(train_ds, batch_size=4, shuffle=True,  num_workers=2)
val_ld   = DataLoader(val_ds,   batch_size=4, shuffle=False, num_workers=2)
test_ld  = DataLoader(test_ds,  batch_size=4, shuffle=False, num_workers=2)

print(f"Train={len(train_ds)}, Val={len(val_ds)}, Test={len(test_ds)}")

# ============================================================
# MODEL ‚Äì U-Net + EfficientNet-B3 encoder
# ============================================================
seg_model = smp.Unet(
    encoder_name="efficientnet-b3",
    encoder_weights="imagenet",
    in_channels=4,
    classes=3,
    activation=None
).to(device)

dice_loss = smp.losses.DiceLoss(mode="multiclass")
ce_loss   = nn.CrossEntropyLoss()

def seg_criterion(pred, target):
    t = target.argmax(1)
    return dice_loss(pred, t) + ce_loss(pred, t)

optimizer_seg = optim.AdamW(seg_model.parameters(), lr=1e-4, weight_decay=1e-5)
scaler_seg    = GradScaler(device.type if device.type == "cuda" else "cpu")

def batch_dice(logits, target, num_classes=3):
    p = logits.argmax(1)
    t = target.argmax(1)
    scores = []
    for c in range(num_classes):
        pc = (p == c)
        tc = (t == c)
        inter = (pc & tc).float().sum()
        union = pc.float().sum() + tc.float().sum()
        if union > 0:
            scores.append((2*inter/union).item())
    return np.mean(scores) if scores else 0.0

# ============================================================
# TRAINING LOOP
# ============================================================
EPOCHS_SEG = 8   # adjust if needed
best_val_dice = 0.0

for epoch in range(1, EPOCHS_SEG+1):
    seg_model.train()
    train_loss = train_dice = 0.0

    for x, y in tqdm(train_ld, desc=f"[SEG Train] {epoch}/{EPOCHS_SEG}"):
        x, y = x.to(device), y.to(device)
        optimizer_seg.zero_grad()
        with autocast(device_type=device.type):
            out  = seg_model(x)
            loss = seg_criterion(out, y)
        scaler_seg.scale(loss).backward()
        scaler_seg.step(optimizer_seg)
        scaler_seg.update()

        train_loss += loss.item()
        train_dice += batch_dice(out, y)

    seg_model.eval()
    val_loss = val_dice = 0.0
    with torch.no_grad():
        for x, y in tqdm(val_ld, desc=f"[SEG Val]   {epoch}/{EPOCHS_SEG}"):
            x, y = x.to(device), y.to(device)
            with autocast(device_type=device.type):
                out  = seg_model(x)
                loss = seg_criterion(out, y)
            val_loss += loss.item()
            val_dice += batch_dice(out, y)

    tl = train_loss / len(train_ld)
    td = train_dice / len(train_ld)
    vl = val_loss   / len(val_ld)
    vd = val_dice   / len(val_ld)

    print(f"\nEpoch {epoch}/{EPOCHS_SEG} ‚Üí Train Dice={td:.4f}, Val Dice={vd:.4f}, Val Loss={vl:.4f}")

    if vd > best_val_dice:
        best_val_dice = vd
        torch.save(seg_model.state_dict(), "best_refuge_unet_effb3.pth")
        print("üî• Saved best segmentation model.")

print("Best validation Dice:", best_val_dice)

# ============================================================
# TEST EVALUATION + vCDR
# ============================================================
seg_model.load_state_dict(torch.load("best_refuge_unet_effb3.pth", map_location=device))
seg_model.eval()

all_preds, all_trues = [], []

with torch.no_grad():
    for x, y in tqdm(test_ld, desc="[SEG Test]"):
        x, y = x.to(device), y.to(device)
        with autocast(device_type=device.type):
            out = seg_model(x)
        pred = out.argmax(1).cpu().numpy().flatten()
        true = y.argmax(1).cpu().numpy().flatten()
        all_preds.append(pred)
        all_trues.append(true)

all_preds = np.concatenate(all_preds)
all_trues = np.concatenate(all_trues)

iou_per_class  = jaccard_score(all_trues, all_preds, labels=[0,1,2],
                               average=None, zero_division=0)
dice_per_class = 2 * iou_per_class / (1 + iou_per_class + 1e-7)
mean_dice      = dice_per_class.mean()

print("\nüîç REFUGE2 SEGMENTATION METRICS")
print("IoU  (BG, Disc, Cup):", iou_per_class)
print("Dice (BG, Disc, Cup):", dice_per_class)
print("Mean Dice:", mean_dice)

# -------- vCDR computation ----------
def compute_vcdr(mask_2d):
    disc = np.where(mask_2d == 1)
    cup  = np.where(mask_2d == 2)
    if len(disc[0]) == 0 or len(cup[0]) == 0:
        return None
    disc_h = disc[0].max() - disc[0].min() + 1
    cup_h  = cup[0].max()  - cup[0].min()  + 1
    if disc_h <= 0:
        return None
    return cup_h / disc_h

vcdr_rows = []
with torch.no_grad():
    for img_path, (x, _) in tqdm(
        list(zip(test_imgs, test_ld)), desc="[vCDR (approx)]", total=len(test_ld)
    ):
        # this zip trick isn't perfect, so recompute correctly:
        break

# Proper vCDR loop (independent of DataLoader zip)
test_single_ld = DataLoader(test_ds, batch_size=1, shuffle=False)
vcdr_values = []
for idx, (x, _) in enumerate(tqdm(test_single_ld, desc="[vCDR per image]")):
    x = x.to(device)
    with torch.no_grad(), autocast(device_type=device.type):
        out = seg_model(x)
    pred = out.argmax(1).cpu().numpy()[0]
    v = compute_vcdr(pred)
    vcdr_values.append({
        "image_path": test_imgs[idx],
        "vCDR_pred": v
    })

cdr_df = pd.DataFrame(vcdr_values)
cdr_df.to_csv("refuge2_test_vcdr.csv", index=False)
print("\nSaved vCDR table to 'refuge2_test_vcdr.csv'")
display(cdr_df.head())

# Quick visualization of a few predictions
def show_examples(dataset, model, n=3):
    model.eval()
    idxs = np.random.choice(len(dataset), min(n, len(dataset)), replace=False)
    plt.figure(figsize=(12, 4*n))
    for i, idx in enumerate(idxs):
        img_t, mask_t = dataset[idx]
        img_np = img_t[:3].numpy().transpose(1,2,0)  # RGB only for display
        mask_true = mask_t.argmax(0).numpy()
        with torch.no_grad(), autocast(device_type=device.type):
            out = model(img_t.unsqueeze(0).to(device))
        mask_pred = out.argmax(1).cpu().numpy()[0]

        plt.subplot(n,3,i*3+1); plt.imshow(img_np); plt.title("Input"); plt.axis("off")
        plt.subplot(n,3,i*3+2); plt.imshow(mask_true, cmap="gray", vmin=0, vmax=2); plt.title("GT"); plt.axis("off")
        plt.subplot(n,3,i*3+3); plt.imshow(mask_pred, cmap="gray", vmin=0, vmax=2); plt.title("Pred"); plt.axis("off")
    plt.tight_layout()
    plt.show()

show_examples(test_ds, seg_model, n=4)
print("‚úÖ Notebook 1 done: segmentation + vCDR CSV.")
```

```python
# ==========================================
# CLASSIFICATION MODEL: EfficientNet-B3
# ==========================================
import os
import numpy as np
import torch
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
from tqdm import tqdm

ACRIMA_ROOT = "/kaggle/input/acrima"  # YOUR Actual folder name

class ACRIMA_Dataset(Dataset):
    def __init__(self, folder, label, transform=None):
        self.paths = [os.path.join(folder, f) for f in os.listdir(folder)]
        self.label = label
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, i):
        img = Image.open(self.paths[i]).convert("RGB")
        if self.transform: img = self.transform(img)
        return img, self.label, self.paths[i]

# -------------------------------
# 1. Load dataset
# -------------------------------
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
])

test_non_glau = os.path.join(ACRIMA_ROOT, "test/Non Glaucoma")
test_glau = os.path.join(ACRIMA_ROOT, "test/Glaucoma")

ds_non = ACRIMA_Dataset(test_non_glau, 0, transform)
ds_glau = ACRIMA_Dataset(test_glau, 1, transform)

test_ds_acrima = ds_non + ds_glau
test_loader = DataLoader(test_ds_acrima, batch_size=32, shuffle=False)

# -------------------------------
# 2. EfficientNet-B3 Classifier
# -------------------------------
model = models.efficientnet_b3(weights="IMAGENET1K_V1")
model.classifier[1] = nn.Linear(1536, 1)
model = model.to("cuda")
model.eval()

# -------------------------------
# 3. Run inference
# -------------------------------
all_rows = []

with torch.no_grad():
    for imgs, labels, paths in tqdm(test_loader):
        imgs = imgs.to("cuda")
        logits = model(imgs).squeeze()
        probs = torch.sigmoid(logits).cpu().numpy()

        for p, lbl, prob in zip(paths, labels, probs):
            all_rows.append([p, lbl, prob])

# -------------------------------
# 4. Save ACRIMA results
# -------------------------------
acrima_df = pd.DataFrame(all_rows, columns=["image", "true_label", "pred_prob"])
acrima_df.to_csv("acrima_test_results.csv", index=False)

print("\nSaved ‚Üí acrima_test_results.csv")
```

```python
!pip install timm --quiet

# ============================================================
# üìå IMPORTS
# ============================================================
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve
)
import seaborn as sns
import timm
from PIL import Image
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================================
# üìå PATHS ‚Äì MODIFY IF NEEDED
# ============================================================
ACRIMA_ROOT = "/kaggle/input/acrima"

TRAIN_DIR = os.path.join(ACRIMA_ROOT, "train")
TEST_DIR  = os.path.join(ACRIMA_ROOT, "test")

# ============================================================
# üìå DATASET CLASS
# ============================================================
class ACRIMADataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]

        if self.transform:
            img = self.transform(img)

        return img, label

# ============================================================
# üìå LOAD DATA
# ============================================================
def load_acrima_split(split_dir):
    glaucoma_dir = os.path.join(split_dir, "Glaucoma")
    normal_dir   = os.path.join(split_dir, "Non Glaucoma")

    glaucoma_imgs = [os.path.join(glaucoma_dir, x) for x in os.listdir(glaucoma_dir)]
    normal_imgs   = [os.path.join(normal_dir, x) for x in os.listdir(normal_dir)]

    all_paths = glaucoma_imgs + normal_imgs
    labels = [1]*len(glaucoma_imgs) + [0]*len(normal_imgs)

    return all_paths, labels

train_paths, train_labels = load_acrima_split(TRAIN_DIR)
test_paths, test_labels   = load_acrima_split(TEST_DIR)

print(f"Training samples: {len(train_paths)}")
print(f"Test samples: {len(test_paths)}")

# ============================================================
# üìå TRANSFORMS
# ============================================================
train_tfms = transforms.Compose([
    transforms.Resize((300,300)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(0.2,0.2,0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225]),
])

test_tfms = transforms.Compose([
    transforms.Resize((300,300)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225]),
])

train_ds = ACRIMADataset(train_paths, train_labels, transform=train_tfms)
test_ds  = ACRIMADataset(test_paths, test_labels, transform=test_tfms)

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True, num_workers=2)
test_loader  = DataLoader(test_ds, batch_size=16, shuffle=False, num_workers=2)

# ============================================================
# üìå MODEL: EfficientNet-B3
# ============================================================
model = timm.create_model("efficientnet_b3", pretrained=True, num_classes=2)
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# ============================================================
# üìå TRAINING LOOP
# ============================================================
EPOCHS = 5
for epoch in range(1, EPOCHS+1):
    model.train()
    total_loss = 0

    for imgs, labels in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS}"):
        imgs, labels = imgs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch} ‚Äî Loss: {total_loss/len(train_loader):.4f}")

torch.save(model.state_dict(), "efficientnet_b3_acrima.pth")
print("\nModel saved as efficientnet_b3_acrima.pth")

# ============================================================
# üìå TESTING / EVALUATION
# ============================================================
model.eval()
y_true = []
y_pred = []
y_prob = []

with torch.no_grad():
    for imgs, labels in tqdm(test_loader, desc="Testing"):
        imgs = imgs.to(device)
        outputs = model(imgs)

        probs = torch.softmax(outputs, dim=1)[:,1].cpu().numpy()
        preds = np.argmax(outputs.cpu().numpy(), axis=1)

        y_true.extend(labels.numpy())
        y_pred.extend(preds)
        y_prob.extend(probs)

# ============================================================
# üìå METRICS
# ============================================================
print("\nüìå CLASSIFICATION REPORT:")
print(classification_report(y_true, y_pred, target_names=["Non-Glaucoma","Glaucoma"]))

print("ROC AUC:", roc_auc_score(y_true, y_prob))

# ============================================================
# üìå CONFUSION MATRIX
# ============================================================
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
            xticklabels=["NG","G"], yticklabels=["NG","G"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# ============================================================
# üìå ROC CURVE
# ============================================================
fpr, tpr, _ = roc_curve(y_true, y_prob)
plt.plot(fpr, tpr, label="EffB3 (AUC = %.4f)" % roc_auc_score(y_true, y_prob))
plt.plot([0,1],[0,1],"k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# ============================================================
# üìå SAVE PREDICTIONS
# ============================================================
df_out = pd.DataFrame({
    "image": test_paths,
    "true_label": y_true,
    "pred_label": y_pred,
    "glaucoma_prob": y_prob
})
df_out.to_csv("acrima_classification_results.csv", index=False)
print("Saved predictions ‚Üí acrima_classification_results.csv")
```
